#### DBN Code: the primary key for future dataframe merge.
The DBN is going to be the key to how we combine all of this data into one dataframe. Therefore every table must have the exact same column. One abberations is data['hs_directory'] which has a 'dbn' column but not a 'DBN'. Thats an easy fix. The other abberation is the class size data, which has the components of the DBN column but not a DBN itself. I will have to apply a function to one of the columns to create this.

```python
# Fix hs_directory
data['hs_directory']['DBN'] = data['hs_directory']['dbn']
```

```python
data['hs_directory'].head()
```

```python
# DBN = two digit csd + school code. Apply a function to csd to ensure it's 2 digits and padded if under 10
# then concatenate the string with the school code and assign to DBN
# csd currently int64 - will need to be converted to string
def pad_csd(csd):
    if csd < 10:
        csd = str(0)+str(csd)
        return csd
    else:
        csd=str(csd)
        return csd
```

```python
#apply to CSD column to create padded CSD
data['class_size']['CSD'] = data['class_size']['CSD'].apply(pad_csd)
```

```python
#Check...
data['class_size'].head()
```

```python
# Check data types, CSD now object
data['class_size'].info()
```

```python
# Create DBN from relevant fields
data['class_size']['DBN'] = data['class_size']['CSD']+data['class_size']['SCHOOL CODE']
```

To capture a geographic component for analysis, the 'Locations 1' field in HS directory dataset contains latitude and longitude. This will be very helpful in conducting analysis and making visualizaitons later. Ill use a regular expression apply to clean 'Location 1' and extract the lat and long for each row.

```python
data['hs_directory'].head()
```

```python
import re
```

```python
# How to use cleaning methods to isolate just the string of the latitude?
coord = re.findall("\(.+\)", "1110 Boston Road\nBronx, NY 10456\n(40.8276026690005, -73.90447525699966)")
spl = str(coord).split(',')
latitude = spl[0][3:]
latitude
```

```python
# Function to find the latitude
def latitude(string):
    coord = re.findall("\(.+\)", string)
    spl = str(coord).split(',')
    latitude = spl[0][3:]
    return latitude
```

```python
# Apply the function to the series to create string latitude
data['hs_directory']['lat'] = data['hs_directory']['Location 1'].apply(latitude)
```

```python
# Display to check work
data['hs_directory'].head()
```

```python
# How to use cleaning methods to isolate just the string of the longitude?
coord = re.findall("\(.+\)", "1110 Boston Road\nBronx, NY 10456\n(40.8276026690005, -73.90447525699966)")
spl = str(coord).split(',')
lon = spl[1][:-3]
lon
```

```python
# Function to find the longitude
def longitude(string):
    coord = re.findall("\(.+\)", string)
    spl = str(coord).split(',')
    lon = spl[1][:-3]
    return lon
```

```python
# Apply to create new column with string longitude
data['hs_directory']['lon'] = data['hs_directory']['Location 1'].apply(longitude)
```

### Combining the dataframes into one dataset:

Before we merge our data, we'll need to decide on the merge strategy we want to use. We'll be using the pandas pandas.DataFrame.merge() function, which supports four types of joins -- left, right, inner, and outer. Each of these join types dictates how pandas combines the rows.

We'll be using the DBN column to identify matching rows across data sets. In other words, the values in that column will help us know which row from the first data set to combine with which row in the second data set.

There may be DBN values that exist in one data set but not in another. This is partly because the data is from different years. Each data set also has inconsistencies in terms of how it was gathered. Human error (and other types of errors) may also play a role. Therefore, we may not find matches for the DBN values in sat_results in all of the other data sets, and other data sets may have DBN values that don't exist in sat_results.

We'll merge two data sets at a time. For example, we'll merge sat_results and hs_directory, then merge the result with ap_2010, then merge the result of that with class_size. We'll continue combining data sets in this way until we've merged all of them. Afterwards, we'll have roughly the same number of rows, but each row will have columns from all of the data sets.

Depending on the strategy we choose, we may preserve rows at the expense of having more missing column data, or minimize missing data at the expense of having fewer rows. Choosing a merge strategy is an important decision; it's worth thinking about your data carefully, and what trade-offs you're willing to make.

Because this project is concerned with determing demographic factors that correlate with SAT score, we'll want to preserve as many rows as possible from sat_results while minimizing null values.

This means that we may need to use different merge strategies with different data sets. Some of the data sets have a lot of missing DBN values. This makes a left join more appropriate, because we don't want to lose too many rows when we merge. If we did an inner join, we would lose the data for many high schools.

Some data sets have DBN values that are almost identical to those in sat_results. Those data sets also have information we need to keep. Most of our analysis would be impossible if a significant number of rows was missing from demographics, for example. Therefore, we'll do an inner join to avoid missing data in these columns.

* Both the ap_2010 and the graduation data sets have many missing DBN values, so we'll use a left join when we merge the sat_results data set with them. Because we're using a left join, our final dataframe will have all of the same DBN values as the original sat_results dataframe.

```python
# list of frames to iterate over and view DBN columns for each
all_frames = ['sat_results','ap_2010','graduation','class_size','demographics','survey','hs_directory']
```

```python
# iterating ove frames to see how many DBN values each frame has
for frame in all_frames:
    print(data[frame].describe(include=['O']))
    
```

```python
combined = data['sat_results']
```

```python
combined = combined.merge(data['ap_2010'],how='left',on='DBN')
combined.head()
```

```python
combined = combined.merge(data['graduation'],how='left',on='DBN')
combined.head()
```

```python
combined.shape
```

```python
#combining all remaining dataframes using inner join. printing shape after each operation executes to track the # rows
frames = ['class_size','demographics','survey','hs_directory']
for frame in frames:
    combined = combined.merge(data[frame],how='inner',on='DBN')
    print(combined.shape)
```

```python
combined.head()
```

#### Missing Data:
There is now many cells with NaN values. This is because we merged with left joins at first, understanding that we preferred missing values over no values at all. 
Now, we must fill the missing data with values that will allow us to perform correlations and other mathematical operations. For this, we will fill NaN's with the column averages. 

We can fill in missing data in pandas using the pandas.DataFrame.fillna() method. This method will replace any missing values in a dataframe with the values we specify. We can compute the mean of every column using the pandas.DataFrame.mean() method. If we pass the results of the df.mean() method into the df.fillna() method, pandas will fill in the missing values in each column with the mean of that column.

We should fill any NaN or null values that remain after the initial replacement with the value 0. We can do this by passing 0 into the df.fillna() method.

```python
# Fill NaN's with means
combined = combined.fillna(combined.mean())
combined.head()
```

```python
# Some NaN's remain because there were no means (object data type columns) or all other values were NaN and the mean was NaN
# Fill remaining with 0
combined = combined.fillna(0)
```

```python
combined.head()
```

### Visualizing ELL percent by School District

In order to explore this relationship, we'll want to map out ell_percent by school district. The map will show us which areas of the city have a lot of English language learners. We can do this utilizing the lat & lon fields that we isolated earlier, paired with the matplotlib basemap package. The first step to reach this goal is to create a map object that shows the locations of all schools in our dataset. Second, we will use different colors to show schools with a high percentage of ELL studends.

1) Create map object with locations of all schools on the map.
In order to plot coordinates using basemap, we need to:
* Convert the pandas series containing the latitude and longitude coordinates to lists using the pandas.Series.tolist() method
* Make a scatterplot using the longitudes and latitudes with the scatter() method on the Basemap object.
* Show the plot using the pyplot.show() method.

```python
from mpl_toolkits.basemap import Basemap
m = Basemap(
    projection='merc', 
    llcrnrlat=40.496044, 
    urcrnrlat=40.915256, 
    llcrnrlon=-74.255735, 
    urcrnrlon=-73.700272,
    resolution='i'
)

m.drawmapboundary(fill_color='#85A6D9')
m.drawcoastlines(color='#6D5F47', linewidth=.4)
m.drawrivers(color='#6D5F47', linewidth=.4)

# Convert latitude & longitude to a list
latitude = combined['lat'].tolist()
longitude = combined['lon'].tolist()
m.scatter(longitude,latitude,s=20,zorder=2,latlon=True)
plt.show()
```

2) Use colors to show schools with high percentages of ELL students. 

The scatter method has a 'c' parameter that serves to adjust the colors of each point on the plot. Passing a range of values to this parameter will result in different shades of the color being displayed. Using the "summer" colormap (cmap) , if we pass c=combined['ell_percent'] into the parameter, the method would shade a school with a high ell_percent yellow, and a school with a low ell_percent green.


```python
m = Basemap(
    projection='merc', 
    llcrnrlat=40.496044, 
    urcrnrlat=40.915256, 
    llcrnrlon=-74.255735, 
    urcrnrlon=-73.700272,
    resolution='i'
)

m.drawmapboundary(fill_color='#85A6D9')
m.drawcoastlines(color='#6D5F47', linewidth=.4)
m.drawrivers(color='#6D5F47', linewidth=.4)

# Convert latitude & longitude to a list
latitude = combined['lat'].tolist()
longitude = combined['lon'].tolist()
m.scatter(longitude,latitude,s=10,zorder=2,latlon=True,c=combined['ell_percent'],cmap='summer')
plt.show()
```

Unfortunately, due to the number of schools, it's hard to interpret the map - there are so many points. It looks like uptown Manhattan and parts of Queens have a higher ell_percent, but we can't be sure. One way to make very granular statistics easier to read is to aggregate them. In this case, we can aggregate by district, which will enable us to plot ell_percent district-by-district instead of school-by-school.

We can use the pandas groupby method to complete this aggregation.

```python
# Mean of ell_percent grouped by district
districts = combined.groupby('school_dist', as_index=False).mean()
districts.head()
```

Now that we have the average ell percent of all schools in each district, we can plot the same basemap plot as before with the data aggregated by district. whats more, we also have the average latitude and longitude of every school in each district as well, which we can use as a proxy for the center of each district. By increasing the size of the points on the scatter plot, we can simplify our graph and communicate more directly the percentage of ell students across all the school districts of New York City. 

```python
m = Basemap(
    projection='merc', 
    llcrnrlat=40.496044, 
    urcrnrlat=40.915256, 
    llcrnrlon=-74.255735, 
    urcrnrlon=-73.700272,
    resolution='i'
)

m.drawmapboundary(fill_color='#85A6D9')
m.drawcoastlines(color='#6D5F47', linewidth=.4)
m.drawrivers(color='#6D5F47', linewidth=.4)

# Convert latitude & longitude to a list - using lat and lon from 'districts' now instead of 'combined'
latitude = districts['lat'].tolist()
longitude = districts['lon'].tolist()
# Increasing size of the points from 10 to 50, same color scheme
m.scatter(longitude,latitude,s=50,zorder=2,latlon=True,c=districts['ell_percent'],cmap='summer')
plt.show()
```

### Investigating the correlation between SAT scores and AP Exam performance

In the U.S., high school students take Advanced Placement (AP) exams to earn college credit. There are AP exams for many different subjects.

It makes sense that the number of students at a school who took AP exams would be highly correlated with the school's SAT scores. Let's explore this relationship. Because total_enrollment is highly correlated with sat_score, we don't want to bias our results. Instead, we'll look at the percentage of students in each school who took at least one AP exam.

```python
# Calculating the percentage of students in each school that took at least 1 AP Exam
combined['ap_per'] = combined['AP Test Takers ']/combined['total_enrollment']
combined.head()
```

```python
# Plotting AP percentage against SAT score
combined.plot.scatter(x='ap_per',y='sat_score')
```

There seems to be a general upward trend in the SAT score average when plotted against the percentage of students that took at least 1 AP test. The exception is due to an almost straight line at about 1200 (y-axis reading) that extends out all the way to the right edge of the plot. This is due to the filled values that we made back in the ealier data cleaning steps, which was necessary because there were about 100 missing schools in the AP_2010 dataset. Nevertheless, the general correlation is positive.

Let's explore the collection of schools with high average SAT scores & that have more than half of all students taking at least one AP exam:

```python
# Filtering for the high performing schools above. 
combined[(combined['ap_per']>.5)&(combined['sat_score']>1600)][['SCHOOL NAME','sat_score','ap_per']]
```

Once again, these three high performing schools are in or near Manhattan, are very competitive for entry, and are billed as college preparatory schools.
